```python
# read in as grayscale

img = cv2.imread('DATA/crossword.jpg',0)

plt.imshow(img,cmap='gray')
```

```python
# a function to show the image larger

def show_pic(img):
    fig = plt.figure(figsize=(15,15))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
show_pic(img)
```

```python
# there is a lot of greyscale info we don't need

# let's adjust the the threshold value

ret,th1 = cv2.threshold(img,180,255,cv2.THRESH_BINARY)
show_pic(th1)
```

```python
# lets try adaptive threshold instead

th2 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,8)
show_pic(th2)
```

```python
# now let's blend the two images (adaptive and threshold value)

blended = cv2.addWeighted(src1=th1,alpha=0.6,src2=th2,beta=0.4,gamma=0)
show_pic(blended)
```
