```python
import cv2
import numpy as np

import matplotlib.pyplot as plt
```

```python
blank_img = np.zeros(shape=(512,512,3),dtype=np.int16)
```

```python
blank_img.shape
```

```python
plt.imshow(blank_img)
```

```python
#draw a rectangle on the image
cv2.rectangle(blank_img,pt1=(384,10),pt2=(500,150),color=(0,255,0),thickness=5)
```

```python
plt.imshow(blank_img)
```

```python
#add a second rectangle
cv2.rectangle(blank_img,pt1=(200,200),pt2=(300,300),color=(0,0,255),thickness=10)
```

```python
plt.imshow(blank_img)
```

```python
#draw a circle
cv2.circle(img=blank_img,center=(100,100),radius=50,color=(255,0,0),thickness=8)
plt.imshow(blank_img)
```

```python
#draw a filled circle
cv2.circle(img=blank_img,center=(400,400),radius=50,color=(255,0,0),thickness=-1)
plt.imshow(blank_img)
```

```python
#draw a line
cv2.line(blank_img,pt1=(0,0),pt2=(512,512),color=(102,255,255),thickness=5)
plt.imshow(blank_img)
```

```python
# draw some text on the image
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(blank_img,text='Kristin',org=(10,500),fontFace=font,fontScale=4,color=(255,255,255),thickness=3,lineType=cv2.LINE_AA)
plt.imshow(blank_img)
```

```python
# create a new blank image
blank_img2 = np.zeros(shape=(512,512,3),dtype=np.int32)

# make an array of vertices with x and y coordinates
vertices = np.array([ [100,300],[200,200],[400,300],[200,400] ],dtype=np.int32)

# need to reshape to 3 dimensions
pts = vertices.reshape((-1,1,2))
```

```python
# draw a custom polygon with the points
cv2.polylines(blank_img2,[pts],isClosed=True,color=(0,255,0),thickness=5)
plt.imshow(blank_img2)
```
